```python
# horizontal stacked bar chart, with the x-axis being the country, and the y-axis being export to each country 2015-2023
import plotly.graph_objects as go

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3']
values1 = [10, 20, 30]
values2 = [15, 10, 5]

# Create figure
fig = go.Figure()

# Add bar for each category
fig.add_trace(go.Bar(
    x=categories,
    y=values1,
    name='Value 1',
    marker_color='indianred'
))

fig.add_trace(go.Bar(
    x=categories,
    y=values2,
    name='Value 2',
    marker_color='lightsalmon'
))

# Change the bar mode to stack
fig.update_layout(barmode='stack')

# Annotations
annotations = []

for i, category in enumerate(categories):
    total_value = values1[i] + values2[i]
    annotations.append(dict(xref='x', yref='y',
                            x=category, y=total_value,
                            text=f'{total_value}%',
                            showarrow=False,
                            font=dict(family='Arial', size=14, color='black'),
                            align='center'))

fig.update_layout(annotations=annotations)

# Show the figure
fig.show()
```
