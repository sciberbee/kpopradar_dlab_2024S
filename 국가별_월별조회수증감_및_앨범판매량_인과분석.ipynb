```python
def show_monthly_youtube_graph(artist, country):
    artist_df = allcountries_views_df[(allcountries_views_df['artist_name'] == artist) & (allcountries_views_df['country'] == country)]
    artist_df['date'] = pd.to_datetime(artist_df['date'], format='%Y%m%d')
    artist_monthly_sum = artist_df.groupby(pd.Grouper(key='date', freq='M')).sum()
    artist_youtube_monthly_sum = artist_monthly_sum['inc']
    plt.figure(figsize=(12, 6))
    plt.plot(artist_youtube_monthly_sum.index, artist_youtube_monthly_sum.values, color='orange', label=artist)
    plt.xlabel('Date')
    plt.ylabel('Views')
    plt.title('Monthly Views of ' + artist + ' in ' + country)
    plt.xticks(rotation=90)
    plt.ticklabel_format(style='plain', axis='y')  # Display y-axis labels in plain format
    plt.legend()
    plt.show()

show_monthly_youtube_graph('BLACKPINK', 'Indonesia')
```

```python
def show_monthly_youtube_graph(country, start, end):
    country_youtube_df = allcountries_views_df[(allcountries_views_df['country'] == country)]
    country_youtube_df['date'] = pd.to_datetime(country_youtube_df['date'], format='%Y%m%d')
    country_monthly_sum = country_youtube_df.groupby(pd.Grouper(key='date', freq='M')).sum()
    country_monthly_sum = country_monthly_sum[(country_monthly_sum.index >= start) & (country_monthly_sum.index <= end)]
    country_monthly_sum.index = country_monthly_sum.index.to_period('M').to_timestamp()
    country_youtube_monthly_sum = country_monthly_sum['inc']
    
    country_album_df = country_album_sales_df[country_album_sales_df['country'] == country]
    country_album_df['date'] = pd.to_datetime(country_album_df['date'], format='%Y-%m-%d')
    country_album_df = country_album_df[(country_album_df['date'] >= start) & (country_album_df['date'] <= end)]
    
    fig, ax1 = plt.subplots(figsize=(12, 8))  # Create the figure and the first axis

    # Plot the first graph
    ax1.plot(country_youtube_monthly_sum.index, country_youtube_monthly_sum.values, color = 'tab:red', marker = 'o', linestyle = '-', label = "YouTube Viewcount Data Points")  # 'r-' is for red solid line
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Monthly YouTube views increment', color='tab:red')  # Set the label for the y-axis
    ax1.set_ylim(country_youtube_monthly_sum.max() * 0.0, country_youtube_monthly_sum.max() * 1.1)  # Set the limits for the y-axis, 
    ###
    # #여기 바꿔가면서 봐야함
    ###
    ax1.tick_params(axis='y', labelcolor='tab:red')  # Set the tick parameters for the y-axis

    # Create a second y-axis for the second graph
    ax2 = ax1.twinx()  
    ax2.plot(country_album_df['date'], country_album_df['export'], color = 'tab:blue', marker = '^', linestyle = '-', label = "Album Sales Data Points")  # 'g-' is for green solid line
    ax2.set_ylabel('Monthly Album sales data (1000 USD)', color='tab:blue')  # Set the label for the y-axis
    ax2.set_ylim(0, country_album_df['export'].max() * 1.1)  # Set the limits for the y-axis
    ax2.tick_params(axis='y', labelcolor='tab:blue')  # Set the tick parameters for the y-axis

    # Show the plot
    plt.title('Monthly YouTube Views and Album Sales of ' + country + ", from " + start + " to " + end)
    plt.legend()
    plt.show()
```

```python
def show_two_graph_and_regression(country, start, end):
    country_youtube_df = allcountries_views_df[(allcountries_views_df['country'] == country)]
    country_youtube_df['date'] = pd.to_datetime(country_youtube_df['date'], format='%Y%m%d')
    country_monthly_sum = country_youtube_df.groupby(pd.Grouper(key='date', freq='M')).sum()
    country_monthly_sum = country_monthly_sum[(country_monthly_sum.index >= start) & (country_monthly_sum.index <= end)]
    country_monthly_sum.index = country_monthly_sum.index.to_period('M').to_timestamp()
    country_youtube_monthly_sum = country_monthly_sum['inc']
    
    country_album_df = country_album_sales_df[country_album_sales_df['country'] == country]
    country_album_df['date'] = pd.to_datetime(country_album_df['date'], format='%Y-%m-%d')
    country_album_df = country_album_df[(country_album_df['date'] >= start) & (country_album_df['date'] <= end)]
    
    fig, ax1 = plt.subplots(figsize=(12, 8))  # Create the figure and the first axis

    # Plot the first graph
    ax1.plot(country_youtube_monthly_sum.index, country_youtube_monthly_sum.values, color = 'tab:red', marker = 'o', linestyle = '-', label = "YouTube Viewcount Data Points")  # 'r-' is for red solid line
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Monthly YouTube views increment', color='tab:red')  # Set the label for the y-axis
    ax1.set_ylim(country_youtube_monthly_sum.max() * 0.5, country_youtube_monthly_sum.max() * 1.1)  # Set the limits for the y-axis, 
    ###
    # #여기 바꿔가면서 봐야함
    ###
    ax1.tick_params(axis='y', labelcolor='tab:red')  # Set the tick parameters for the y-axis

    # Create a second y-axis for the second graph
    ax2 = ax1.twinx()  
    ax2.plot(country_album_df['date'], country_album_df['export'], color = 'tab:blue', marker = '^', linestyle = '-', label = "Album Sales Data Points")  # 'g-' is for green solid line
    ax2.set_ylabel('Monthly Album sales data (1000 USD)', color='tab:blue')  # Set the label for the y-axis
    ax2.set_ylim(0, country_album_df['export'].max() * 1.1)  # Set the limits for the y-axis
    ax2.tick_params(axis='y', labelcolor='tab:blue')  # Set the tick parameters for the y-axis
    
    # Show the plot
    plt.title('Monthly YouTube Views and Album Sales of ' + country + ", from " + start + " to " + end)
    plt.legend()
    plt.show()
    
    # I. Linearly fit the two data points and plot the regression line in second plot
    # Get the x and y values for the regression line
    x = country_youtube_monthly_sum.values
    y = country_album_df['export'].values
    print("x.shape, y.shape: ", x.shape, y.shape)
    # Calculate the correlation coefficient
    correlation = np.corrcoef(x, y)[0, 1]
    # Calculate the linear regression
    m, b = np.polyfit(x, y, 1)
    # Plot the regression line
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, 'o', label='Original data', color = 'tab:blue')
    plt.plot(x, m*x + b, label='Fitted line', color = 'tab:red')
    plt.xlabel('Monthly YouTube views increment')
    plt.ylabel('Monthly Album sales data (1000 USD)')
    plt.title('Linear Regression, ' + country + ", from " + start + " to " + end)
    plt.legend()
    plt.show()
    # print the correlation coefficient
    print("Correlation Coefficient: ", correlation)
    # print the mean squared error
    print("Mean Squared Error: ", np.mean((m*x + b - y) ** 2))
```

```python
show_two_graph_and_regression('United States', '2021-05-01', '2023-09-30')
```
